# 판다스를 이용하여 데이터와 친해지기

머신러닝 프로젝트의 첫 번째 단계는 데이터와 친해지는 것입니다. 이를 위해 판다스 라이브러리를 사용할 것입니다. 판다스는 과학자들이 데이터를 탐색하고 조작하기 위해 사용하는 주요 도구입니다. 대부분의 사람들은 판다스를 `pd`로 줄여 적습니다. 이는 다음 명령어로 수행합니다.

```python
import pandas as pd
```

판다스 라이브러리의 가장 중요한 부분은 데이터프레임입니다. 데이터프레임은 표(table)라고 생각할 수 있는 형태로 데이터를 저장합니다. 이는 엑셀에서의 시트나 SQL 데이터베이스의 표와 비슷합니다.

판다스는 데이터로 하고 싶어 할 만한 대부분의 작업에 대한 강력한 메소드를 제공합니다.

예를 들어, 우리는 호주 멜버른의 [주택값 데이터](https://www.kaggle.com/dansbecker/melbourne-housing-snapshot)를 살펴볼 것입니다.

예제 (멜버른) 데이터는 파일 경로 **`../../data/melbourne-housing-snapshot/melb_data.csv`** 에 저장되어 있습니다.

다음 명령어를 이용해 데이터를 불러오고 탐색해 봅니다:

```python
# 쉬운 접근을 위해 파일 경로를 변수에 저장합니다
melbourne_file_path = '../../data/melbourne-housing-snapshot/melb_data.csv'
# 데이터를 읽어 melbourne_data라는 이름의 데이터프레임으로 저장합니다
melbourne_data = pd.read_csv(melbourne_file_path) 
# 멜버른 데이터의 요약을 출력합니다
melbourne_data.describe()
```

# 데이터 설명 해석
결과는 원래 데이터의 각 열에 대해 숫자 8개를 표시합니다. 첫 번째 숫자인 **count**는 결측값이 아닌 행의 수를 나타냅니다.

결측값은 여러 가지 이유로 발생합니다. 예를 들어 침실이 하나인 집은 두 번째 침실 크기를 기록하지 않습니다. 결측값에 대한 설명은 이후에 계속하겠습니다.

두 번째 값은 평균을 나타내는 **mean**입니다. 그 아래는 표준편차를 나타내는 **std**로, 값들이 얼마나 흩어져 있는지를 나타냅니다.

**min**, **25%**, **50%**, **75%**, **max** 값을 해석하기 위해 각 행을 오름차순으로 정렬한다고 가정합니다. 첫 번째 값은 최솟값이 됩니다. 4분의 1 지점의 값은 전체 값들의 25%보다 크고 75%보다 작습니다. 이 값이 **25%** 값입니다. 50%와 75% 값도 마찬가지로 구하며, **max**는 최댓값입니다.
