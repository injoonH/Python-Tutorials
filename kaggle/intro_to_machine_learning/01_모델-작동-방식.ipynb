# 인트로
머신러닝 모델의 작동 방식과 사용 방법에 대한 개요부터 살펴보겠습니다. 이전에 통계 모델링이나 머신러닝을 해본 적이 있다면 이는 기본이라고 느낄 수 있습니다. 진행함에 있어 강력한 모델들을 만들어 나갈 것이기에 걱정하지 않으셔도 됩니다.

이 코스에서는 다음 시나리오에 따라 모델을 제작합니다:

당신의 사촌은 부동산 투기로 수백만 달러를 벌었습니다. 그는 당신의 데이터 과학에 대한 관심 때문에 비지니스 파트너가 될 것을 제안합니다. 그는 돈을 투자하고, 당신은 다양한 주택의 가치를 예측하는 모델을 공급하게 될 것입니다.

당신은 사촌에게 과거에 부동산 가치를 어떻게 예측했냐고 물었고, 사촌은 그저 직관이었다고 말합니다. 그러나 더 많은 조사를 통해 그가 과거에 본 집들로부터 가격 패턴을 알아냈고, 그는 그 패턴을 이용하여 새로운 집들에 대한 예측을 했다는 사실을 알아냈습니다.

머신러닝도 같은 방식으로 작동합니다. 우리는 의사 결정 트리라고 불리는 모델부터 다루기 시작할 것입니다. 더 정확한 예측을 하는 화려한 모델들도 있습니다. 그러나 의사 결정 트리는 이해하기 쉬우며, 데이터 과학 분야에서 가장 우수한 모델 중 일부를 위한 기본 구성 요소입니다.

우선은 가장 단순한 의사 결정 트리부터 시작하겠습니다.

![First Decision Trees](http://i.imgur.com/7tsb5b1.png)

위는 집들을 단지 두 개의 범주로 나눕니다. 고려된 주택들의 예상가는 과거 같은 범주에 들었던 주택들의 평균가입니다.

우리는 데이터를 사용하여 주택을 두 그룹으로 나눈 다음 각 그룹의 예상가를 결정합니다. 데이터에서 패턴을 찾는 과정을 **피팅** 혹은 **훈련**이라고 합니다. 모델의 **피팅**하기 위해 사용한 데이터를 **훈련 데이터**라고 합니다.

모델을 피팅하는 과정은 (e.g. 데이터를 둘로 나누는 과정) 너무 복잡하여 설명을 뒤로 미루겠습니다. 모델을 피팅한 후에는 새로운 데이터에 적용하여 주택 가격을 **예측**할 수 있습니다.

# 의사 결정 트리 발전시키기
다음 중 실제 훈련 데이터를 사용했을 때의 결과로 적합한 의사 결정 트리는 어느 것입니까?

![First Decision Trees](http://i.imgur.com/prAjgku.png)

일반적으로 침실이 많은 주택일수록 높은 가격에 팔리는 경향이 있기 때문에 왼쪽의 의사 결정 트리가 (Decision Tree 1) 더 말이 될 것입니다. 이 모델의 가장 큰 단점은 욕실 수, 부지 크기, 위치 등 집값에 영향을 미치는 대부분의 요소를 포착하지 못한다는 것입니다.

더 많은 "분할"이 있는 트리를 사용하여 더 많은 요인을 포착할 수 있습니다. 이들을 "깊은" 트리라고 부릅니다. 주택의 부지 크기를 고려하는 의사 결정 트리는 다음과 같을 것입니다:

![Depth 2 Tree](http://i.imgur.com/R3ywQsR.png)

당신은 어떤 주택이더라도 그의 특성에 맞는 경로를 선택함으로써 집값을 예측할 수 있습니다. 주택의 예상가는 트리의 가장 아래에 있습니다. 예측을 하는, 트리의 가장 아래를 **리프(leaf)**라고 부릅니다.

분할과 리프의 값은 데이터에 따라 결정되므로 이제 당신이 작업할 데이터를 확인할 차례입니다.
